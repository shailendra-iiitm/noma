## Analyze Results
Let's analyze the clustering results:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Compare throughput
throughputs = {
    method: result['metrics']['total_throughput']
    for method, result in results.items()
}

plt.figure(figsize=(10, 6))
plt.bar(throughputs.keys(), throughputs.values())
plt.title('Total Throughput Comparison')
plt.ylabel('Throughput (Mbps)')
plt.show()
```

```python
# Compare NOMA coverage
coverage = {
    method: result['metrics']['noma_coverage']
    for method, result in results.items()
}

plt.figure(figsize=(10, 6))
plt.bar(coverage.keys(), coverage.values())
plt.title('NOMA Coverage Comparison')
plt.ylabel('Coverage (%)')
plt.show()
```
